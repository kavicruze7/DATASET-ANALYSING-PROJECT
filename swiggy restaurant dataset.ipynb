```python
#Data Manipulation
import pandas as pd

#Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>Importing The Dataset

```python
df = pd.read_csv('swiggy.csv')
```

```python
#To known the columns in the dataset 
df.columns
```

```python
#To know shape of the dataset
df.shape
```

```python
#To known the first five rows in the dataset
df.head()
```

```python
#To known the last five rows in the dataset
df.tail()
```

```python
#To know info of the dataset
df.info()
```

```python
#Checking the missing values of the dataset
df.isnull().sum()
```

```python
#changing price dtype[FLOAT] to dtype[INT64]
df['Price'] = df['Price'].astype('int64')
df['Price'].unique()
```

```python
#Getting unique cities from the dataset
df['City'].unique()
```

```python
#Finding which restaurant has maximum in the dataset
df['Restaurant'].max()
```

```python
#Finding which food type has maximum in the dataset
df['Food type'].max()
```

```python
#Describing Statistic Summary
df.describe()
```

```python
#finding total price in chennai using group by
df.groupby(['City','Area'])['Price'].sum()['Chennai'].max()
```

```python
#taking counts for the avg ratings
df['Avg ratings'].value_counts()
```

```python
#finding no.of restro in the cities
Cities = df['City'].value_counts().sort_values(ascending = False)
plt.bar(Cities.index,Cities.values,color = "hotpink")
plt.title("NO.OF.RESTAURANT IN THE CITIES",fontdict={'fontweight':'bold'})
plt.xticks(rotation = 90)
plt.xlabel = "Cities"
plt.ylabel = "Count"
for i,v in enumerate(Cities.values):
    plt.text(i,v,str(v),ha='center')
plt.show()

print("\nThe No.Of.Restaurant in the cities:  ",Cities)

```

```python
#Top 3 City has most of the restaurant
Top_Res = df.groupby("City")["Restaurant"].count().sort_values(ascending=False).head(3)
plt.bar(Top_Res.index,Top_Res.values,color = "yellow")
plt.title("TOP THREE RESTAURANT IN THE CITIES",fontdict={'fontweight':'bold'})
plt.xticks(rotation = 90)
plt.xlabel = "Top_Res"
plt.ylabel = "Count"
for i,v in enumerate(Top_Res.values):
    plt.text(i,v,str(v),ha='center')
plt.show()

print("\nTop 3 Restaurant in the cities:  ",Top_Res)
```

```python
plt.scatter(df['Avg ratings'],df['City'], color = "blue")
plt.show()
```

```python
#Top 10 food types in the dataset
Top_Food = df['Food type'].value_counts().sort_values(ascending = False).head(10)
plt.bar(Top_Food.index,Top_Food.values,color = "orange") 
plt.title("TOP 10 FOOD TYPES",fontdict={'fontweight':'bold'})
plt.xticks(rotation = 'vertical')
plt.xlabel = "Food Type"
plt.ylabel = "Count"
for i,v in enumerate(Top_Food.values):
    plt.text(i,v,str(v),ha='center')
plt.show()

print("TOP 10 FOODS:  ",Top_Food)
```

```python
top_5restro = df.groupby('Restaurant')['Avg ratings'].count().sort_values(ascending=False).head(5)
plt.bar(top_5restro.index,top_5restro.values,color = "violet")
plt.title("TOP 5 RESTAURANT BY AVG RATINGS",fontdict={'fontweight':'bold'})
plt.xticks(rotation = 90)
plt.xlabel = "top_5restro"
plt.ylabel = "Count"
for i,v in enumerate(top_5restro.values):
    plt.text(i,v,str(v),ha='center')
plt.show()

print("\nTop 5 Restaurant by avg ratings:  ",top_5restro)
```

```python
city_price = df.groupby('City')['Price'].mean().sort_values()
least_expensive_city = city_price.head(3)
print(f"The city with the least expensive restaurants is {least_expensive_city}.")
plt.pie(least_expensive_city)
plt.show()
```

```python
chennai_df = df[df['City'] == 'Chennai']
most_popular_restaurant = chennai_df.sort_values('Total ratings', ascending=False)['Restaurant'].index.tolist
print(f"The restaurant in Bangalore with the most number of people visited is {most_popular_restaurant}.")
```

```python
sns.countplot(data = df, x = "City")
plt.xticks(rotation=70)
```

```python
#highest avg rating restro
df.groupby("Restaurant")["Avg ratings"].mean().sort_values(ascending=False).head(1)
```

```python
#highest avg rating city
Top_Rating_City = df.groupby("City")["Avg ratings"].mean().sort_values(ascending=False).head(5)
plt.bar(Top_Rating_City.index,Top_Rating_City.values,color = "violet")
plt.title("TOP 5 AVG RATINGS CITY",fontdict={'fontweight':'bold'})
plt.xticks(rotation = 90)
plt.xlabel = "Top_Rating_City"
plt.ylabel = "Count"
for i,v in enumerate(Top_Rating_City.values):
    plt.text(i,v,str(v),ha='center')
plt.show()

print("\nTop 5 Avg_Ratings City:  ",Top_Rating_City)
```

```python
#highest 3 Total rating area
Top_Rating_Foods = df.groupby("Food type")["Total ratings"].mean().sort_values(ascending=False).head(3)
plt.bar(Top_Rating_Foods.index,Top_Rating_Foods.values,color = "#bcbd22")
plt.title("TOP 3 RATINGS FOODS",fontdict={'fontweight':'bold'})
plt.xticks(rotation = 90)
plt.xlabel = "Top_Rating_Foods"
plt.ylabel = "Count"
for i,v in enumerate(Top_Rating_Foods.values):
    plt.text(i,v,str(v),ha='center')
plt.show()

print("\nTop # Ratings Foods:  ",Top_Rating_Foods)
```
